# Question 20

```python
quest20=pd.read_csv('Cars.csv')
```

```python
quest20
```

```python
# P(MPG>38)
1-stats.norm.cdf(38,quest20.MPG.mean(),quest20.MPG.std())
```

```python
# P(MPG<40)
stats.norm.cdf(40,quest20.MPG.mean(),quest20.MPG.std())
```

```python
# P (20<MPG<50)
stats.norm.cdf(50,quest20.MPG.mean(),quest20.MPG.std())-(1-stats.norm.cdf(20,quest20.MPG.mean(),quest20.MPG.std()))
```
